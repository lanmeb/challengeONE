### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print(f"Faturamento da Loja 1: R${faturamento_loja1:.2f}")
print(f"Faturamento da Loja 2: R${faturamento_loja2:.2f}")
print(f"Faturamento da Loja 3: R${faturamento_loja3:.2f}")
print(f"Faturamento da Loja 4: R${faturamento_loja4:.2f}")

# Faturamento total
faturamento_total = faturamento_loja1 + faturamento_loja2 + faturamento_loja3 + faturamento_loja4
print(f"Faturamento Total: R${faturamento_total:.2f}")
```

```python
import matplotlib.pyplot as plt
# Gráfico
labels = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
valores = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]
percentuais = [(valor / faturamento_total) * 100 for valor in valores]

# Criando o gráfico de pizza
plt.figure(figsize=(4, 4))
plt.pie(valores, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Percentual de Faturamento por Loja')
plt.show()
```

# 2. Vendas por Categoria


```python
print("Loja 1:")
print(loja["Categoria do Produto"].value_counts(), "\n")

print("Loja 2:")
print(loja2["Categoria do Produto"].value_counts(), "\n")

print("Loja 3:")
print(loja3["Categoria do Produto"].value_counts(), "\n")

print("Loja 4:")
print(loja4["Categoria do Produto"].value_counts(), "\n")
```

```python
import matplotlib.pyplot as plt
# Calculando os value_counts para cada loja
dados_lojas = pd.DataFrame({
    "Loja 1": loja["Categoria do Produto"].value_counts(),
    "Loja 2": loja2["Categoria do Produto"].value_counts(),
    "Loja 3": loja3["Categoria do Produto"].value_counts(),
    "Loja 4": loja4["Categoria do Produto"].value_counts()
})
# Gráfico de barras para cada loja
dados_lojas.plot(kind='bar', figsize=(10, 6))

plt.title("Distribuição de Categorias de Produto por Loja")
plt.xlabel("Categoria do Produto")
plt.ylabel("Frequência")
plt.legend(title="Lojas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
# Calculando a média diretamente pela coluna "Avaliação da compra"
media_loja1 = loja["Avaliação da compra"].mean()
media_loja2 = loja2["Avaliação da compra"].mean()
media_loja3 = loja3["Avaliação da compra"].mean()
media_loja4 = loja4["Avaliação da compra"].mean()

# Exibindo as médias

print(f"Média da Loja 1: {media_loja1:.2f}")
print(f"Média da Loja 2: {media_loja2:.2f}")
print(f"Média da Loja 3: {media_loja3:.2f}")
print(f"Média da Loja 4: {media_loja4:.2f}")


# Calculando a média geral entre todas as lojas
media_geral = (media_loja1 + media_loja2 + media_loja3 + media_loja4) / 4
print(f"Média geral entre as lojas: {media_geral:.2f}")
```

```python

# Configurando os dados para o gráfico
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
medias = [media_loja1, media_loja2, media_loja3, media_loja4]

# Criando o gráfico de barras
plt.bar(lojas, medias, color=["blue", "green", "orange", "red"])
plt.title("Média das Avaliações por Loja", fontsize=16)
plt.xlabel("Lojas", fontsize=12)
plt.ylabel("Média de Avaliações", fontsize=12)
plt.ylim(0, 5)  # Escala de avaliações (0 a 5)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Adicionando os valores das médias acima das barras
for i in range(len(medias)):
    plt.text(i, medias[i] + 0.1, f"{medias[i]:.2f}", ha='center', fontsize=10)

# Exibir o gráfico
plt.show()
```

```python
# Configurar a figura para exibir 4 gráficos separados por loja
fig, axes = plt.subplots(2, 2, figsize=(15, 12), sharey=True)

# Loja 1: Gráfico de barras com valores
produtos_menos_loja1 = dados_lojas["Loja 1"].nsmallest(5)
produtos_mais_loja1 = dados_lojas["Loja 1"].nlargest(5)
axes[0, 0].bar(produtos_menos_loja1.index, produtos_menos_loja1, color='red', label="Menos Vendidos")
axes[0, 0].bar(produtos_mais_loja1.index, produtos_mais_loja1, color='blue', label="Mais Vendidos")
axes[0, 0].set_title("Loja 1")
axes[0, 0].set_ylabel("Quantidade Vendida")
axes[0, 0].tick_params(axis='x', rotation=90)

# Adicionar os valores no topo das barras
for i, produto in enumerate(produtos_menos_loja1.index):
    valor = produtos_menos_loja1[produto]
    axes[0, 0].text(i, valor + 0.5, str(valor), ha='center', va='bottom', fontsize=9)
for i, produto in enumerate(produtos_mais_loja1.index):
    # Usar uma nova posição baseada no índice dos mais vendidos
    valor = produtos_mais_loja1[produto]
    eixo_x = produtos_menos_loja1.index.get_loc(produto) if produto in produtos_menos_loja1.index else i + len(produtos_menos_loja1)
    axes[0, 0].text(eixo_x, valor + 0.5, str(valor), ha='center', va='bottom', fontsize=9)


# Loja 2: Gráfico de barras com valores
produtos_menos_loja2 = dados_lojas["Loja 2"].nsmallest(5)
produtos_mais_loja2 = dados_lojas["Loja 2"].nlargest(5)
axes[0, 1].bar(produtos_menos_loja2.index, produtos_menos_loja2, color='red', label="Menos Vendidos")
axes[0, 1].bar(produtos_mais_loja2.index, produtos_mais_loja2, color='blue', label="Mais Vendidos")
axes[0, 1].set_title("Loja 2")
axes[0, 1].tick_params(axis='x', rotation=90)


# Adicionar os valores no topo das barras (menos vendidos)
for i, produto in enumerate(produtos_menos_loja2.index):
    valor = produtos_menos_loja2[produto]
    axes[0, 1].text(i, valor + 0.5, str(valor), ha='center', va='bottom', fontsize=9)

# Adicionar os valores no topo das barras (mais vendidos)
for i, produto in enumerate(produtos_mais_loja2.index):
    # Usar uma nova posição baseada no índice dos mais vendidos
    valor = produtos_mais_loja2[produto]
    eixo_x = produtos_menos_loja2.index.get_loc(produto) if produto in produtos_menos_loja2.index else i + len(produtos_menos_loja2)
    axes[0, 1].text(eixo_x, valor + 0.5, str(valor), ha='center', va='bottom', fontsize=9)


# Loja 3: Gráfico de barras com valores
produtos_menos_loja3 = dados_lojas["Loja 3"].nsmallest(5)
produtos_mais_loja3 = dados_lojas["Loja 3"].nlargest(5)
axes[1, 0].bar(produtos_menos_loja3.index, produtos_menos_loja3, color='red', label="Menos Vendidos")
axes[1, 0].bar(produtos_mais_loja3.index, produtos_mais_loja3, color='blue', label="Mais Vendidos")
axes[1, 0].set_title("Loja 3")
axes[1, 0].set_ylabel("Quantidade Vendida")
axes[1, 0].tick_params(axis='x', rotation=90)

# Adicionar os valores no topo das barras
for i, (produto, valor) in enumerate(produtos_menos_loja3.items()):
    axes[1, 0].text(i, valor + 0.5, str(valor), ha='center', va='bottom', fontsize=9)
for i, produto in enumerate(produtos_mais_loja3.index):
    # Usar uma nova posição baseada no índice dos mais vendidos
    valor = produtos_mais_loja3[produto]
    eixo_x = produtos_menos_loja3.index.get_loc(produto) if produto in produtos_menos_loja3.index else i + len(produtos_menos_loja3)
    axes[1, 0].text(eixo_x, valor + 0.5, str(valor), ha='center', va='bottom', fontsize=9)

# Loja 4: Gráfico de barras com valores
produtos_menos_loja4 = dados_lojas["Loja 4"].nsmallest(5)
produtos_mais_loja4 = dados_lojas["Loja 4"].nlargest(5)
axes[1, 1].bar(produtos_menos_loja4.index, produtos_menos_loja4, color='red', label="Menos Vendidos")
axes[1, 1].bar(produtos_mais_loja4.index, produtos_mais_loja4, color='blue', label="Mais Vendidos")
axes[1, 1].set_title("Loja 4")
axes[1, 1].tick_params(axis='x', rotation=90)

# Adicionar os valores no topo das barras
for i, (produto, valor) in enumerate(produtos_menos_loja4.items()):
    axes[1, 1].text(i, valor + 0.5, str(valor), ha='center', va='bottom', fontsize=9)
for i, produto in enumerate(produtos_mais_loja4.index):
    # Usar uma nova posição baseada no índice dos mais vendidos
    valor = produtos_mais_loja4[produto]
    eixo_x = produtos_menos_loja4.index.get_loc(produto) if produto in produtos_menos_loja4.index else i + len(produtos_menos_loja4)
    axes[1, 1].text(eixo_x, valor + 0.5, str(valor), ha='center', va='bottom', fontsize=9)

# Adicionar legendas para todos os gráficos
for ax in axes.flat:
    ax.legend()


# Ajustar layout para evitar sobreposição
plt.tight_layout()
plt.show()
```

# 5. Frete Médio por Loja


```python
mediaFreteLoja1 = loja['Frete'].mean()
mediaFreteLoja2 = loja2['Frete'].mean()
mediaFreteLoja3 = loja3['Frete'].mean()
mediaFreteLoja4 = loja4['Frete'].mean()

print(f"Média de Frete da Loja 1: R${mediaFreteLoja1:.2f}")
print(f"Média de Frete da Loja 2: R${mediaFreteLoja2:.2f}")
print(f"Média de Frete da Loja 3: R${mediaFreteLoja3:.2f}")
print(f"Média de Frete da Loja 4: R${mediaFreteLoja4:.2f}")

totalFrete = mediaFreteLoja1 + mediaFreteLoja2 + mediaFreteLoja3 + mediaFreteLoja4
print(f"Média de Frete Total: R${totalFrete:.2f}")
```

```python
# Configurando os dados para o gráfico
lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
medias = [mediaFreteLoja1, mediaFreteLoja2, mediaFreteLoja3, mediaFreteLoja4]
# Criando o gráfico de barras
plt.bar(lojas, medias, color=["blue", "green", "orange", "red"])
plt.title("Média de Frete por Loja", fontsize=16)
plt.xlabel("Lojas", fontsize=12)
plt.ylabel("Média de Frete (R$)", fontsize=12)
plt.ylim(0, max(medias) + 5)  # Ajuste a escala do eixo y conforme necessário
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Adicionando os valores acima das barras
for i in range(len(medias)):
    plt.text(i, medias[i] + 0.2, f"R${medias[i]:.2f}", ha='center', fontsize=10)

# Exibindo o gráfico
plt.show()
```
